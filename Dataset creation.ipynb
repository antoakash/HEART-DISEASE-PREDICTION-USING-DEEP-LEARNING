```python
import pydicom
from PIL import Image
import numpy as np
import os

def dicom_to_image(dicom_path, output_path, output_format='PNG'):
    """
    Convert a DICOM file to an image and save it in the specified format.
    """
    # Read the DICOM file
    dicom_data = pydicom.dcmread(dicom_path)
    
    # Get the image pixel data
    image_data = dicom_data.pixel_array
    
    # Normalize the image data (optional but can help with visualization)
    image_data_normalized = np.uint8(image_data / np.max(image_data) * 255)
    
    # Convert the numpy array to a PIL image
    image = Image.fromarray(image_data_normalized)
    
    # Save the image in the desired format (PNG by default)
    image.save(output_path, format=output_format)
    print(f"Image saved as {output_path}")

def convert_images_in_folder(input_folder, output_folder, prefixes=('CINELAX_', 'CINESAX_'), output_format='PNG'):
    """
    Convert all DICOM files inside subfolders starting with the specified prefixes
    and save them in the output folder with continuous numbering.
    """
    # Create output folder if it does not exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # List all subfolders in the input folder
    subfolders = [f for f in os.listdir(input_folder) if os.path.isdir(os.path.join(input_folder, f))]

    # Filter subfolders that start with specified prefixes (CINELAX_ or CINESAX_)
    valid_subfolders = [subfolder for subfolder in subfolders if subfolder.startswith(prefixes)]
    
    # Initialize a global image counter
    image_counter = 4301
    
    # Iterate over each valid subfolder
    for subfolder in valid_subfolders:
        subfolder_path = os.path.join(input_folder, subfolder)
        
        # List all DICOM files in the current subfolder
        dicom_files = [f for f in os.listdir(subfolder_path) if f.endswith('.dcm')]
        
        # Iterate over all DICOM files and convert them
        for dicom_file in dicom_files:
            dicom_path = os.path.join(subfolder_path, dicom_file)
            
            # Format the output image name to include the global image counter
            output_image_name = f"image_{image_counter:03d}.{output_format.lower()}"   # e.g., image_001.png
            output_path = os.path.join(output_folder, output_image_name)
            
            # Convert and save the image
            dicom_to_image(dicom_path, output_path, output_format)
            
            # Increment the global image counter
            image_counter += 1

# Example usage
input_folder = '/Users/akash/Downloads/SCD_IMAGES_05/SCD0004501'  # Parent folder containing subfolders like CINELAX_7, CINESAX_2, etc.
output_folder = '/Users/akash/Desktop/MRI DATASET PREDICTION/Normal'  # Folder where you want to save the .png or .jpg images
convert_images_in_folder(input_folder, output_folder, prefixes=('CINELAX_', 'CINESAX_','AORTA_'), output_format='PNG')
```
